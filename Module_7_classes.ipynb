```python
# Class variable
# now we will define class variable
# instance variable class variable
# class and object
# instance variable are write in seperate method and class varible are write in class
class Computer:
    # class variable these area class variable name 
    name = "Dell"
    

    def __init__(self , ram , ssd , processor):
        # instance varible we are define in special or init method
        self.com_ram = ram
        self.com_ssd = ssd
        self.processor = "I5"
        print("hi")
        # these are all is instance variable
    # this is instance method    
    def features(self):
        # these are using instance varible this is instance method
        print("Ram of computer is :" , self.com_ram)
        print("SSD of computer is :" , self.com_ssd)
        print("Processor is :" , self.processor)
        
    # the above two are instance methods
    
    # here we defining class method 
    @classmethod
    def company_Name(cls):
        # to use the class instance will be class method
        print("Company name is :" , cls.name)
    
    
    # Static method
    @staticmethod
    def price(price):
        
        print("Price of the computer is:" , price)


com1 = Computer(16 , 256 , "i7")
com2 = Computer(32 , 512 , "i7")
com1.processor = "I7"

com1.features()
com2.features()
com2.company_Name()
com2.price("45K")
```

```python
# Class and Instance Variables


# Class for Dog
class Dog:
   
    # Class Variable
    animal = 'dog'            
   
    # The init method or constructor
    def __init__(self, breed, color):
     
        # Instance Variable    
        self.breed = breed
        self.color = color       
    
# Objects of Dog class
Rodger = Dog("Pug", "brown")
Buzo = Dog("Bulldog", "black")
 
print('Rodger details:')  
print('Rodger is a', Rodger.animal)
print('Breed: ', Rodger.breed)
print('Color: ', Rodger.color)
 
print('\nBuzo details:')  
print('Buzo is a', Buzo.animal)
print('Breed: ', Buzo.breed)
print('Color: ', Buzo.color)
 
# Class variables can be accessed using class
# name also
print("\nAccessing class variable using class name")
print(Dog.animal)
```

```python
# Constructors in Python

# Example of default constructor : 

class GeekforGeeks:
 
    # default constructor
    def __init__(self):
        self.geek = "GeekforGeeks"
 
    # a method for printing data members
    def print_Geek(self):
        print(self.geek)
 
 
# creating object of the class
obj = GeekforGeeks()
 
# calling the instance method using the object obj
obj.print_Geek()
```

```python
# Example of the parameterized constructor : 


class Addition:
    first = 0
    second = 0
    answer = 0
     
    # parameterized constructor
    def __init__(self, f, s):
        self.first = f
        self.second = s
     
    def display(self):
        print("First number = " + str(self.first))
        print("Second number = " + str(self.second))
        print("Addition of two numbers = " + str(self.answer))
 
    def calculate(self):
        self.answer = self.first + self.second
 
# creating object of the class
# this will invoke parameterized constructor
obj = Addition(1000, 2000)
 
# perform Addition
obj.calculate()
 
# display result
obj.display()
```
